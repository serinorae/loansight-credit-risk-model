```python
import pandas as pd
import os 
file_path = '../data/raw/'
file_name = 'accepted_2007_to_2018Q4.csv'
full_path = os.path.join(file_path, file_name)
df = pd.read_csv(full_path, low_memory=False)
print("Dataset loaded successfully")
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns}")
df.info()
df['loan_status'].value_counts()
```

```python
df_sample = df.sample(frac=0.1, random_state=42)
pd.set_option('display.max_rows', None)
null_info = pd.DataFrame({
    'dtype': df_sample.dtypes,
    'not_null_count': df_sample.notnull().sum(),
    'null_count': df_sample.isnull().sum(),
    'null_pct': (df_sample.isnull().sum() / len(df_sample) * 100).round(2)
})
print(null_info)
```

```python
df = df[["loan_amnt", "annual_inc", "int_rate", "dti", "term", "loan_status"]]
null_details = pd.DataFrame({
    'dtype': df.dtypes,
    'null_count': df.isnull().sum(),
    'null_pct': (df.isnull().sum() / len(df) * 100).round(2)
})
print(null_details)
df = df.dropna(subset=["dti"])
df = df[df["loan_status"].isin(["Fully Paid", "Charged Off", "Default"])]
df['loan_status'].value_counts()
df["default"] = df["loan_status"].apply(
    lambda x: 1 if x.lower() in ["charged off", "default"] else 0
)
df.to_csv("../data/processed/cleaned_loans.csv", index=False)
```
